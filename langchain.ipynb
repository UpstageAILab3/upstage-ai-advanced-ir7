```python
import wikipedia

def wiki_single_search(query, lang='en'):
    wikipedia.set_lang(lang)
    page = wikipedia.page(query)

    # text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 200)
    # docs = text_splitter.split_text(text)
    return page
```

```python
import wikipedia
from wikipedia.exceptions import DisambiguationError, PageError

def wiki_find_docs(query, lang='en'):
    wikipedia.set_lang(lang)
    pages = wikipedia.search(query)

    docs = []
    for page in set(pages):
        try:
            doc = wikipedia.page(page)
            docs.append(doc)
        except DisambiguationError as e:
            print(f"'{page}'는 여러 의미를 가질 수 있습니다. 첫 번째 옵션을 선택합니다.")
            if e.options:
                try:
                    doc = wikipedia.page(e.options[0])
                    docs.append(doc)
                except PageError:
                    print(f"'{e.options[0]}'에 대한 페이지를 찾을 수 없습니다.")
        except PageError:
            print(f"'{page}'에 대한 페이지를 찾을 수 없습니다.")
    
    return docs
```

```python

query = "public education"
page = wiki_single_search(query) 

print(page.title)
print(page.content)
print("\n--------------------------------------------------------------\n")

docs = wiki_find_docs(query)
for doc in docs:
    print(doc.title)
    print(doc.content)
    print()
```

# Tavily 사용해보자

```python
from tavily import TavilyClient

from dotenv import load_dotenv
import os
from langdetect import detect

# API KEY 설정
load_dotenv('/upstage-ai-advanced-ir7/.env')
tavily_api_key = os.getenv("TAVILY_API_KEY")
os.environ["TAVILY_API_KEY"] = tavily_api_key

def get_tavily_data(query):
    tavily_api_key = os.getenv("TAVILY_API_KEY")
    client = TavilyClient(api_key=tavily_api_key)

    results = client.search(query = query, max_results = 10)

    ko_results = [result for result in results['results'] if detect(result['content']) == 'ko']

    return ko_results
```

```python
get_tavily_data("드미트리")
```
